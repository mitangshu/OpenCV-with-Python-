```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
heigth = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
output = cv2.VideoWriter('myvideo.mkv', cv2.VideoWriter_fourcc(*'DIVX'), 20, (width, heigth) )
```

# Record the Video


```python
%matplotlib inline
while True:
  ret, frame = cap.read()

  #operations
  output.write(frame)

  #display
  # If you want gray frame
  #gray = cv2.CvtColor(frame,cv2.COLOR_BGR2GRAY)
  #cv2.imshow(frame, gray)

  cv2.imshow('frame', frame)
  # never use x to get out of the frame use 27 or any other key like esc
  if cv2.waitKey(5) & 0xff==27:
    break

cap.release()
output.release()
cv2.destroyAllWindows()
```

# Draw  on live Videos

```python
import cv2

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

x = width // 2
y = height // 2

w = width // 4
h = height // 4

while cap.isOpened():
    
    ret, frame = cap.read()
    
    # Rectangle Frame
    
    cv2.rectangle(frame, (x,y), (x+w , y+h), color = (255,0,255), thickness = 5)

    cv2.imshow('frame', frame)
        
    if cv2.waitKey(3) & 0xff==27:
        break

cap.release()

cv2.destroyAllWindows()
        
```
